```python
df_eurovision_finales["percentage_points"] = round((df_eurovision_finales["total_points"]/df_eurovision_finales["total_points_per_year"])*100, 2)
```

```python
df_eurovision_losers = df_eurovision_finales.groupby("year")["rank"].max().reset_index()
```

```python
for year in df_eurovision_finales["year"]:
    if year in(df_eurovision_losers["year"]):
        if df_eurovision_losers["rank"] == df_eurovision_finales["rank"]:
            df_eurovision_finales["losers"] = True
        else:
            df_eurovision_finales["losers"] = False
```
